```python
import random
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

Dataset

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Crear individuo

```python
def crear_individuo():
    n_estimators = random.randint(10, 100)
    max_depth = random.randint(1, 10)
    return [n_estimators, max_depth]
```

Crear población

```python
def crear_poblacion(tamano):
    return [crear_individuo() for _ in range(tamano)]
```

FUNCIÓN DE FITNESS

```python
def fitness(individuo):
    n_estimators, max_depth = individuo

    modelo = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth
    )

    modelo.fit(X_train, y_train)
    return modelo.score(X_test, y_test)
```

Selección (torneo simple)

```python
def seleccion(poblacion):
    a = random.choice(poblacion)
    b = random.choice(poblacion)
    return a if fitness(a) > fitness(b) else b
```

Cruzamiento (mezcla directa)

```python
def cruzar(padre1, padre2):
    hijo = [
        random.choice([padre1[0], padre2[0]]),
        random.choice([padre1[1], padre2[1]])
    ]
    return hijo
```

Mutación

```python
def mutar(individuo, prob=0.2):
    if random.random() < prob:
        individuo[0] = random.randint(10, 100)

    if random.random() < prob:
        individuo[1] = random.randint(1, 10)

    return individuo
```

AG

```python
# Crear población inicial
poblacion = crear_poblacion(10)

# Guardar mejor global
mejor_global = None
mejor_fitness = 0

# Ciclo de generaciones (TERMINACIÓN)
for generacion in range(10):

    nueva_poblacion = []

    for _ in range(len(poblacion)):

        # SELECCIÓN
        padre1 = seleccion(poblacion)
        padre2 = seleccion(poblacion)

        # CRUZAMIENTO
        hijo = cruzar(padre1, padre2)

        # MUTACIÓN
        hijo = mutar(hijo)

        nueva_poblacion.append(hijo)

    # Reemplazo de población
    poblacion = nueva_poblacion

    # Evaluar mejor de la generación
    mejor = max(poblacion, key=fitness)
    fit = fitness(mejor)

    # Guardar mejor global
    if fit > mejor_fitness:
        mejor_fitness = fit
        mejor_global = mejor

    print(f"Gen {generacion}: Mejor = {mejor}, Fitness = {fit}")

# Resultado final
print("\nMejor GLOBAL:", mejor_global)
print("Mejor Accuracy:", mejor_fitness)
```
