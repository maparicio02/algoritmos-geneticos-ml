```python
import random
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

Dataset

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Crear individuo

```python
def crear_individuo(n_features):
    return [random.randint(0, 1) for _ in range(n_features)]
```

Crear población

```python
def crear_poblacion(tamano, n_features):
    return [crear_individuo(n_features) for _ in range(tamano)]
```

FUNCIÓN DE FITNESS

```python
def fitness(individuo):
    # seleccionar features activas
    indices = [i for i in range(len(individuo)) if individuo[i] == 1]

    # evitar individuo vacío
    if len(indices) == 0:
        return 0

    X_train_sel = X_train[:, indices]
    X_test_sel = X_test[:, indices]

    modelo = RandomForestClassifier()
    modelo.fit(X_train_sel, y_train)

    return modelo.score(X_test_sel, y_test)
```

Selección (torneo simple)

```python
def seleccion(poblacion):
    a = random.choice(poblacion)
    b = random.choice(poblacion)

    return a if fitness(a) > fitness(b) else b
```

Cruzamiento (cruce de un punto)

```python
def cruzar(padre1, padre2):
    punto = random.randint(1, len(padre1)-1)
    hijo = padre1[:punto] + padre2[punto:]
    return hijo
```

Mutación

```python
def mutar(individuo, prob=0.1):
    for i in range(len(individuo)):
        if random.random() < prob:
            individuo[i] = 1 - individuo[i]
    return individuo
```

AG

```python
poblacion = crear_poblacion(10, 4)

for generacion in range(10):
    nueva_poblacion = []

    for _ in range(len(poblacion)):
        padre1 = seleccion(poblacion)
        padre2 = seleccion(poblacion)

        hijo = cruzar(padre1, padre2)
        hijo = mutar(hijo)

        nueva_poblacion.append(hijo)

    poblacion = nueva_poblacion

    # mostrar mejor de la generación
    mejor = max(poblacion, key=fitness)
    print(f"Generación {generacion}: Mejor = {mejor}, Fitness = {fitness(mejor)}")
```

```python
mejor = max(poblacion, key=fitness)
print("\nMejor solución final:", mejor)
print("Accuracy:", fitness(mejor))
```
