```python
import random
```

```python
X = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

y = [0, 0, 0, 1]
```

```python
def crear_individuo():
    return [
        random.uniform(-2, 2),  # w1
        random.uniform(-2, 2),  # w2
        random.uniform(-2, 2)   # bias
    ]
```

```python
def crear_poblacion(tamano):
    return [crear_individuo() for _ in range(tamano)]

poblacion = crear_poblacion(20)
```

```python
def red_neuronal(x, individuo):
    w1, w2, b = individuo
    return 1 if (x[0]*w1 + x[1]*w2 + b) > 0 else 0
```

```python
def fitness(individuo):
    correctos = 0

    for i in range(len(X)):
        pred = red_neuronal(X[i], individuo)
        if pred == y[i]:
            correctos += 1

    return correctos / len(X)
```

```python
def seleccion(poblacion):
    a = random.choice(poblacion)
    b = random.choice(poblacion)
    return a if fitness(a) > fitness(b) else b
```

```python
def cruzar(p1, p2):
    return [
        (p1[0] + p2[0]) / 2,
        (p1[1] + p2[1]) / 2,
        (p1[2] + p2[2]) / 2
    ]
```

```python
def mutar(individuo, prob=0.2):
    for i in range(len(individuo)):
        if random.random() < prob:
            individuo[i] += random.uniform(-1, 1)
    return individuo
```

```python
mejor_global = None
mejor_fitness = 0

for gen in range(30):

    nueva_poblacion = []

    for _ in range(len(poblacion)):

        p1 = seleccion(poblacion)
        p2 = seleccion(poblacion)

        hijo = cruzar(p1, p2)
        hijo = mutar(hijo)

        nueva_poblacion.append(hijo)

    poblacion = nueva_poblacion

    mejor = max(poblacion, key=fitness)
    fit = fitness(mejor)

    if fit > mejor_fitness:
        mejor_fitness = fit
        mejor_global = mejor

    print(f"Gen {gen}: Mejor = {mejor}, Fitness = {fit}")

print("\nMejor solución:", mejor_global)
print("Accuracy:", mejor_fitness)
```
